## Data Cleaning

Here we clean the dataset by handling missing values, fixing data types, and removing unnecessary columns.

```python
df=df.dropna(subset=['artist_name'])
df
```

```python
df.isnull().sum()
```

```python
df["artist_genres"] = df["artist_genres"].fillna(df["artist_genres"].mode()[0])
```

```python
df.isnull().sum()
```

## Exploratory Data Analysis (EDA)

EDA helps uncover patterns and trends within the data.

```python
plt.figure(figsize=(8,6))
sns.histplot(df["track_popularity"], bins=30, kde=True)
plt.title("Distribution of Track Popularity")
plt.show()
```

```python
df.head()
```

```python
df.groupby('artist_name')['artist_followers'].max().sort_values(ascending=False).head(10).reset_index()
```

```python
df.sort_values('track_popularity', ascending=False).head(10)[['track_name', 'track_popularity']]
```

## Data Visualization

Visualizations are used to better understand insights from the Spotify data.

```python
df.artist_name.value_counts().head(10).plot(kind='bar',figsize=(10,5))
plt.title("Top 10 Artists by Number of Tracks")
plt.show()
```

```python
df['explicit'].value_counts().plot(kind='pie', autopct='%.2f%%',colors=['red','skyblue'],shadow=True,labels=["Non-Explicit","Explicit"])
plt.title("Explicit vs Non-Explicit Tracks")
plt.show()
```

```python
explode=[0.04,0.08,0.1,]
df['album_type'].value_counts().plot(kind='pie', explode=explode, autopct='%.2f%%',colors=['yellow','green','pink'],shadow=True)
plt.title("Album Type")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df,
    x="track_duration_min",
    y="track_popularity",
    hue="explicit"
)
plt.title("Track Duration vs Popularity")
plt.show()
```

```python
df.head()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(data=df,x="explicit",y="track_popularity")
plt.title("Track Popularity by Explicit ")
plt.show()
```
